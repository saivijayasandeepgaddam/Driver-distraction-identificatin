```python
train_image = []
i = 0
def create_train_image():
    for category in classes:
        path = os.path.join(train1,category)
        class_num = classes.index(category)
        
        for image in os.listdir(path):
            image_array = cv2.imread(os.path.join(path,image),cv2.IMREAD_COLOR)
            grey_scale = cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB)
            modf_image = cv2.resize(grey_scale,(image_size,image_size))
            train_image.append([modf_image,class_num])
```

```python
create_train_image()
```

```python
len(train_image)
```

```python
random.shuffle(train_image)
x = []
y = []
```

```python
for features, label in train_image:
    x.append(features)
    y.append(label)
```

```python
len(x)
```

```python
len(x)
```

```python
len(y)
```

```python
x[0].shape
```

```python
x = np.array(x).reshape(-1,image_size,image_size,3)
x[0].shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=96)
```

```python
import gc
gc.collect()
```

```python
from tensorflow.keras import utils
```

```python
Y_train = utils.to_categorical(y_train,num_classes=10)
Y_test = utils.to_categorical(y_test,num_classes=10)
```

```python
from tensorflow.keras import datasets, layers, models, callbacks
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten, Activation, BatchNormalization
```

```python

model = models.Sequential()
#layer1
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(240,240,3)))
model.add(BatchNormalization())
model.add(Conv2D(32,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization(axis = 3))
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
model.add(Dropout(0.3))
#layer 2
model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization(axis = 3))
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
model.add(Dropout(0.3))
#layer 3
model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization(axis = 3))
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
model.add(Dropout(0.5))
#output layer
model.add(Flatten())
model.add(Dense(units = 512,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units = 128,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
#epocs=7;10,15
#epocs are number of passes of data trough the neural net work. which fits the curve less epocs curve will be linear
#more epoces curve over fits 
#batch size bigger the batch size better will be the result but proccesing may crash
#lower batch size will be linear path
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
callback = [callbacks.EarlyStopping(monitor='val_accuracy',patience=5)]
batch_size = 70
n_epochs = 15
```

```python
results = model.fit(x_train,Y_train,
                    batch_size=batch_size,epochs=n_epochs,
                    verbose=1,
                    validation_data=(x_test,Y_test),
                    callbacks=callback)
```

```python
pred= model.predict_classes(x_test) 
```

```python
pred
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test, pred))
```
